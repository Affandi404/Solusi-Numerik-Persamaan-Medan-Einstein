```python
import math
import numpy as np
import matplotlib.pyplot as plt

r0 = 2.8
rs = 2.0
rn = r0
dr = 0.01
ar0 = 1-(rs/r0)
br0 = 1/ar0
arm = ar0
brm = br0
#variabel untuk grafik
X, Ya, Yb = [], [], []
YaEu, YbEu, YaHe, YbHe = [], [], [], []
YeEa, YeEb, YeHa, YeHb = [], [], [], []

for i in range(1,5000):
    r = rn + dr
    rn = r

    #analitical solutions
    ar = 1-(rs/rn)
    br = 1/ar
    #diff equations
    b1r = (brm/rn)*(1-brm)
    a1r = (1/rn)-(1/(brm*rn))

    #euler method
    breu = br
    areu = ar
    brp1eu = breu + dr*b1r
    breu = brp1eu
    arp1eu = areu + dr*a1r
    areu = arp1eu

    #Heun Method
    brhe = br
    arhe = ar
    b1rhe = (brhe/r)*(1-brhe)
    a1rhe = (1/r)-(1/(br*r))
    brp1he = brhe + dr*(b1r+b1rhe)/2
    brhe = brp1he
    arp1he = arhe + dr*(a1r+a1rhe)/2
    arhe = arp1he
    
    #numerical error
    #eruler method
    eEbr = (brp1eu-br)/br
    eEar = (arp1eu-ar)/ar
    #Heun Method
    eHbr = (brp1he-br)/br
    eHar = (arp1he-ar)/ar

    #Save your data
    X.append(1/rn)
    Ya.append(ar)
    Yb.append(br)
    YaEu.append(arp1eu)
    YbEu.append(brp1eu)
    YaHe.append(arp1he)
    YbHe.append(brp1he)
    YeEa.append(eEar)
    YeEb.append(eEbr)
    YeHa.append(eHar)
    YeHb.append(eHbr)

#tampilkan plot A(r) secara numerik
plt.plot(X,YaEu,'g+')
plt.plot(X,YaHe,'bx')
plt.plot(X,Ya,'k')
plt.legend(["Euler",
            "Heun",
            "Analitik"], loc=1)
plt.xlabel('1/r (m⁻¹) ', fontsize=12)
plt.ylabel('Elemen Tensor Metrik A(r)', fontsize=12)
plt.title('Schwarzschild Metrik', fontsize=17)
plt.show()
plt.savefig("out1.png")
print("A(r).Curv =", np.min(ar))
print("A(r).Curv.EU =", np.min(arp1eu))
print("A(r).Curv.He =", np.min(arp1he))

#tampilkan plot B(r) secara numerik
plt.plot(X,YbEu,'g+')
plt.plot(X,YbHe,'bx')
plt.plot(X,Yb,'k')
plt.legend(["Euler",
            "Heun",
            "Analitik"], loc=2)
plt.xlabel('1/r (m⁻¹) ', fontsize=12)
plt.ylabel('Elemen Tensor Metrik B(r)', fontsize=12)
plt.title('Metrik Schwarzschild', fontsize=17)
plt.show()
plt.savefig("out2.png")
print("B(r).Curv =", np.min(br))
print("B(r).Curv.EU =", np.min(brp1eu))
print("B(r).Curv.He =", np.min(brp1he))

#tampilkan plot Error A(r) secara numerik
plt.plot(X,YeEa,'r--')
plt.plot(X,YeHa,'b:')
plt.legend(["Euler",
            "Heun"], loc=2)
plt.xlabel('1/r (m⁻¹) ', fontsize=12)
plt.ylabel('Erorr Relatif A(r)', fontsize=12)
plt.title('Akurasi Numerik', fontsize=17)
plt.show()
plt.savefig("out3.png")
print("ΔFD.Max =", np.max(eEar))
print("ΔHE.Max =", np.max(eHar))

#tampilkan plot Error B(r) secara numerik
plt.plot(X,YeEb,'r--')
plt.plot(X,YeHb,'b:')
plt.legend(["Euler",
            "Heun"], loc=1)
plt.xlabel('1/r (m⁻¹) ', fontsize=12)
plt.ylabel('Erorr Relatif B(r)', fontsize=12)
plt.title('Akurasi Numerik', fontsize=17)
plt.show()
plt.savefig("out4.png")
print("ΔFD.Max =", np.max(eEbr))
print("ΔHE.Max =", np.max(eHbr))
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

r0 = 2.8
rs = 2.0
rm = r0
dr = 0.01

#variabel untuk grafik
X, Ya, Yb = [], [], []
YaEu, YaHe, YbHe, YbFd, YbNm = [], [], [], [], []
YeEa, YeHa, YeHb, YeFb, YeNb = [], [], [], [], []

for i in range(1,5000):
    r = rm + dr
    rm = r
    rp = rm + 2*dr
    rm = rp

    #analitical solutions
    km = 1-(rs/(4*rm))
    kp = 1+(rs/(4*rm))
    kmn = 1-(rs/(4*r))
    kpn = 1+(rs/(4*r))
    kmp = 1-(rs/(4*rp))
    kpp = 1+(rs/(4*rp))
    arm = km**2/kp**2
    brm = kp**4
    ar = kmn**2/kp**2
    br = kpn**4
    brp = kpp**4
    
    #differensial tensor ricci
    b1rm = -(rs/rm)*kp**3
    b2rm = ((3*b1rm**2)/4*brm)-((2*b1rm)/rm)
    a1rm = (-arm*b1rm*(rm*b1rm+4*brm))/(2*rm*b1rm*brm+4*(brm**2))
    
    #Fd and euler method
    arp1eu = arm + dr*a1rm
    arm = arp1eu
    brp1fd = 2*br - brm + b2rm*dr**2
    br = brm
    brm = brp1fd
    
    #Heun Method
    brhe = br
    arhe = ar
    
    b1r = -(rs/r)*kp**3
    b2r = ((3*b1r**2)/4*br)-((2*b1r)/r)
    a1r = (-ar*b1r*(r*b1r+4*br))/(2*r*b1r*br+4*(brm**2))
    
    b1rp = -(rs/rp)*kpp**3
    b2rp = ((3*b1rp**2)/4*brp)-((2*b1rp)/rp)

    arp1he = arhe + dr*(a1r+a1rm)/2
    arhe = arp1he
    brp1he = 2*brhe - brm + dr*(b2rm-2*b2r+b2rp)/2
    brhe = brm
    brm = brp1he
    
    #Numrov method
    brn = br
    brp1nm = 2*brn - brm + ((b2rm+10*b2r+b2rp)/12)*dr**2
    brn = brm
    brm = brp1nm

    #numerical error
    #eruler and FD method
    eFbr = (brp1fd-br)/br
    eEar = (arp1eu-ar)/ar
    #Heun Method
    eHbr = (brp1he-br)/br
    eHar = (arp1he-ar)/ar
    #Numerov method
    eNbr = (brp1nm-br)/br
    
    #Save your data
    X.append(1/r)
    Ya.append(ar)
    Yb.append(br)
    YaEu.append(arp1eu)
    YbFd.append(brp1fd)
    YaHe.append(arp1he)
    YbHe.append(brp1he)
    YbNm.append(brp1nm)
    YeEa.append(eEar)
    YeFb.append(eFbr)
    YeHa.append(eHar)
    YeHb.append(eHbr)
    YeNb.append(eNbr)

#tampilkan plot A(r) secara numerik
plt.plot(X,YaHe,'bx')
plt.plot(X,YaEu,'g+')
plt.plot(X,Ya,'k')
plt.legend(["Heun",
            "Euler",
            "Analitik"], loc=1)
plt.xlabel('1/r (m⁻¹) ', fontsize=12)
plt.ylabel('Elemen Tensor Metrik A(r)', fontsize=12)
plt.title('Metrik Isotropik', fontsize=17)
plt.show()
plt.savefig("out5.png")
print("A(r).Curv =", np.min(ar))
print("A(r).Curv.EU =", np.min(arp1eu))
print("A(r).Curv.He =", np.min(arp1he))

#tampilkan plot B(r) secara numerik
plt.plot(X,YbNm,'ro')
plt.plot(X,YbFd,'g+')
plt.plot(X,YbHe,'bx')
plt.plot(X,Yb,'k')
plt.legend(["Numerov",
            "Beda Hingga",
            "Heun",
            "Analitik"], loc=2)
plt.xlabel('1/r (m⁻¹) ', fontsize=12)
plt.ylabel('Elemen Tensor Metrik B(r)', fontsize=12)
plt.title('Metrik Isotropik ', fontsize=17)
plt.show()
plt.savefig("out6.png")
print("B(r).Curv =", np.min(br))
print("B(r).Curv.EU =", np.min(brp1fd))
print("B(r).Curv.He =", np.min(brp1he))
print("B(r).Curv.Nm =", np.min(brp1nm))

#tampilkan plot Error A(r) secara numerik
plt.plot(X,YeEa,'r--')
plt.plot(X,YeHa,'b:')
plt.legend(["Euler",
            "Heun"], loc=2)
plt.xlabel('1/r (m⁻¹) ', fontsize=12)
plt.ylabel('Erorr Relatif A(r)', fontsize=12)
plt.title('Akurasi Numerik', fontsize=17)
plt.show()
plt.savefig("out7.png")
print("ΔEU.Max =", np.max(eEar))
print("ΔHE.Max =", np.max(eHar))

#tampilkan plot Error B(r) secara numerik
plt.plot(X,YeFb,'r--')
plt.plot(X,YeHb,'b:')
plt.plot(X,YeNb,'g:')
plt.legend(["Beda Hingga",
            "Heun",
            "Numerov"], loc=3)
plt.xlabel('1/r (m⁻¹) ', fontsize=12)
plt.ylabel('Erorr Relatif B(r)', fontsize=12)
plt.title('Akurasi Numerik', fontsize=17)
plt.show()
plt.savefig("out8.png")
print("ΔFD.Max =", np.max(eFbr))
print("ΔHE.Max =", np.max(eHbr))
print("ΔNM.Max =", np.max(eNbr))
```
